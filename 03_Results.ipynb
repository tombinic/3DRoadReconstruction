```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from scipy.interpolate import Rbf
```

# Deep Learning approaches

```python


world_points = pd.read_csv('/content/drive/My Drive/IACV/Point_clouds/F_RCNN_Deep_video_curve_short.csv').values

world_points = [pt for pt in world_points if (pt[2] > 15)]

x_world = [pt[0] for pt in world_points]
y_world = [pt[1] for pt in world_points]
z_world = [pt[2] for pt in world_points]

indices = np.arange(len(x_world))
normalized_indices = indices / max(indices)


fig = go.Figure(data=[go.Scatter3d(
    x=x_world, y=y_world, z=z_world,
    mode='markers',
    marker=dict(
        size=5,
        color=normalized_indices,
        colorscale='Bluered',
        colorbar=dict(title='time instant')
    )
)])

fig.update_layout(
    title="F_RCNN_Deep_video_curve_long",
    scene=dict(
        xaxis=dict(
            title='X Coordinate',
        ),
        yaxis=dict(
            title='Y Coordinate',
        ),
        zaxis=dict(
            title='Z Coordinate',
        )
    ),
    autosize=False,
    width=700,
    height=500,
    margin=dict(l=65, r=50, b=65, t=90)
)

fig.show()
```
